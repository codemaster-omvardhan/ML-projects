### Option 1: Mounting Google Drive (Recommended for most users)

This allows you to access files stored in your Google Drive directly from your Colab notebook. The files will persist even if your Colab session ends.

First, mount your Google Drive:

```python
from google.colab import drive
drive.mount('/content/drive')
```

Once mounted, your Google Drive contents will be available under `/content/drive/MyDrive/`. You can then access your datasets by specifying their full path within your Google Drive, like this (replace `YourFolder` and `your_dataset.csv` with your actual path and filename):

```python
import pandas as pd

# Example: Read a CSV file from Google Drive
# Make sure to update the path to where your file is located in your Google Drive
file_path_in_drive = '/content/drive/MyDrive/YourFolder/your_dataset.csv'

try:
    df_drive = pd.read_csv(file_path_in_drive)
    print(f'Successfully loaded {file_path_in_drive} into a pandas DataFrame.')
    display(df_drive.head())
except FileNotFoundError:
    print(f"Error: The file '{file_path_in_drive}' was not found. Please check the path in your Google Drive.")
except Exception as e:
    print(f"An error occurred: {e}")
```
